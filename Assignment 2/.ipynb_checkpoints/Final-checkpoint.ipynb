```python
import time
stime = time.time()

import struct as st
import numpy as np

filename = {'images' : 'train-images-idx3-ubyte' ,'labels' : 'train-labels-idx1-ubyte'}

labels_array = np.array([])

data_types = {
        0x08: ('ubyte', 'B', 1),
        0x09: ('byte', 'b', 1),
        0x0B: ('>i2', 'h', 2),
        0x0C: ('>i4', 'i', 4),
        0x0D: ('>f4', 'f', 4),
        0x0E: ('>f8', 'd', 8)}

for name in filename.keys():
    if name == 'images':
        imagesfile = open(filename[name],'rb')
    if name == 'labels':
        labelsfile = open(filename[name],'rb')

#reading magic number
imagesfile.seek(0)
magic = st.unpack('>4B',imagesfile.read(4))

print(magic)
if(magic[0] and magic[1])or(magic[2] not in data_types):
    raise ValueError("File Format not correct")

nDim = magic[3]
print("Data is ",nDim,"-D")

#offset = 0004 for number of images
#offset = 0008 for number of rows
#offset = 0012 for number of columns
#32-bit integer (32 bits = 4 bytes)
imagesfile.seek(4)
nImg = st.unpack('>I',imagesfile.read(4))[0] #num of images/labels
nR = st.unpack('>I',imagesfile.read(4))[0] #num of rows
nC = st.unpack('>I',imagesfile.read(4))[0] #num of columns
nBytes = nImg*nR*nC
labelsfile.seek(8) #Since no. of items = no. of images and is already read
print("no. of images :: ",nImg)
print("no. of rows :: ",nR)
print("no. of columns :: ",nC)

#Read all data bytes at once and then reshape
images_array = 255 - np.asarray(st.unpack('>'+'B'*nBytes,imagesfile.read(nBytes))).reshape((nImg,nR*nC))
labels_array = np.asarray(st.unpack('>'+'B'*nImg,labelsfile.read(nImg))).reshape((nImg,1))

print("Time of execution : %s seconds" % str(time.time()-stime))
```

```python


class Neural_Network(object):
    def __init__(self,inputSize,outputSize,hiddenSize):
    #parameters
        self.inputSize = inputSize
        self.outputSize = outputSize
        self.hiddenSize = hiddenSize
        self.weightsItoH=np.random.randn(self.inputSize, self.hiddenSize)
        self.weightsHtoO=np.random.randn(self.hiddenSize, self.outputSize)
        self.xj=np.array([])
        self.learningRate=0.5
    
    def sigmoid(self,x):
        return 1/(1+np.exp(-x))

    def sigmoidDerivative(self,s):
        return s*(1-s)
    
    def forwardPropagation(self):
        print("In FP")
        ItoHDotProduct=np.dot(images_array,self.weightsItoH)
        self.xj=self.sigmoid(ItoHDotProduct)
        HtoODotProduct=np.dot(self.xj,self.weightsHtoO)
        return self.sigmoid(HtoODotProduct)
    
    def backPropagation(self,expectedOutput):
        error=labels_array-expectedOutput
        sensitivityHtoO=error*self.sigmoidDerivative(expectedOutput)
        deltaWij=self.learningRate*self.xj.T.dot(sensitivityHtoO)
        
        errorItoH=sensitivityHtoO.dot(self.weightsHtoO.T)
        sensitivityItoH=errorItoH*self.sigmoidDerivative(self.xj)
        deltaWjk=self.learningRate*images_array.T.dot(sensitivityItoH)
        
        self.weightsHtoO+=deltaWij
        self.weightsItoH+=deltaWjk
        print("error",error,"errorItoH",errorItoH"self.weightsHtoO",self.weightsHtoO,"self.weightsItoH",self.weightsItoH,sep=" : ")
        
    def train(self):
        expectedOutput=self.forwardPropagation()
        self.backPropagation(expectedOutput)
        
    def predict(self):
        print(str(np.mean(np.square(labels_array - self.forwardPropagation()))))
        
        
```

```python
# print(images_array)
# print(labels_array)
nn=Neural_Network(784,10,700)
# print(nn.forwardPropagation(images_array.reshape(60000,784)))
# nn.sigmoidDerivative(2)



for i in range(50):
    nn.train()
    nn.predict()
    print(i)

```
